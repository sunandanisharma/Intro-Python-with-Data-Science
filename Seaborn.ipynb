```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
#Generate random number
number =  np.random.randn(100)
print(number)
```

```python
# draw our first distplot diagram
sns.distplot(number)
```

```python
#change the deault color to green
sns.distplot(number, color='green')
```

```python
#label our x values 
varible_x = pd.Series(number, name='Variable X')
sns.distplot(varible_x)
```

```python
#change label to appear in y axis 
sns.distplot(varible_x,vertical=True)
```

```python
# Display only kde line 
sns.distplot(varible_x, hist=False)
```

```python
# Display only histogram diagram
sns.distplot(varible_x,kde=False)
```

```python
# Draw heatmp diagram
num = np.random.rand(12,15)
sns.heatmap(num) 
#dark color are closer to zero and light color are closer to one
```

```python
# Display number for each rectangle 
sns.heatmap(num,annot=True)
```

```python
#Change range values by using vmax and vmin 
sns.heatmap(num, vmin=0, vmax=2)
```

```python
# Draw violin plot for tips column 
sns.violinplot(x=number)
```

```python
# Draw a strip plot 
sns.stripplot(x=number)
```

```python
#lOAD IRIS DATASET 
iris = sns.load_dataset('iris')
```

```python
#draw pairgrid for iris dataset with scatter plot 
x = sns.PairGrid(iris)
x= x.map(plt.scatter)
```

```python
#draw pairgrid for iris dataset and show only the diagonal part 
x = sns.PairGrid(iris)
x= x.map_diag(plt.hist)
```
